```python


import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
sns.set(style="whitegrid")

# read the data
data = pd.read_csv("../python_psf_external_19.csv")
```

```python
# print every column
for col in data.columns:
    print(col)
# we are interested in "use optional type hinting:When developing in Python, how often do you …?"
```

```python
# Save interestring columns in variables to avoid typos
hints_column = "use optional type hinting:When developing in Python, how often do you …?"
domains_column = "What do you use Python for the most?"
```

```python

hints_data = data[hints_column]
hints_values = hints_data.dropna().unique()
# hints_values = hints_data.unique()
# Get possible values, without "nan"
print(hints_values)
```

```python
# Get the number of occurences for each value
hints_data.value_counts(dropna=False)
```

```python
# Compute the percentage people didn't answer the question
hints_data.isna().sum() / hints_data.shape[0] * 100
```

```python
# Compute the use of type hints for each domain of programmation
domains_values = data[domains_column].unique()
print(domains_values)
```

```python
f, ax = plt.subplots(figsize=(20, 10))
data.groupby([domains_column, hints_column]).size().unstack().sort_values(domains_column, ascending=True).plot.bar(ax=ax, stacked=True)
```
